```python
# Array of Edges (directed) [start,end]
from collections import defaultdict, deque

n = 8
A = [[0, 1], [1, 2], [0, 3], [3, 4], [3, 6], [3, 7], [4, 2], [4, 5], [5, 2]]

# Adj matrix (not used freq.)
M = [[0] * n for _ in range(n)]
for u, v in A:
    M[u][v] = 1
    # if graph is undirected -> opp. connection
    # M[v][u] = 1

# Adj. List

D = defaultdict(list)

for u, v in A:
    D[u].append(v)
    # if graph is undirected -> opp. connection
    #  M[v].append(u)


# DFS recursive
def dfs_recursive(node):
    print(node)
    for nei in D[node]:
        if nei not in seen:
            dfs_recursive(nei)
    seen.add(node)

start = 0
seen = set()
seen.add(start)
print("--- DFS recursive --- ")
dfs_recursive(start)

# DFS iterative
seen.clear()
seen.add(start)
stack = [start]
print("--- DFS iterative --- ")
while stack:
    node = stack.pop()
    print(node)
    for nei in D[node]:
        if nei not in seen:
            stack.append(nei)
    seen.add(node)

print("--- BFS - Queue --- ")
seen.clear()
seen.add(start)
queue = deque([start])
print("--- DFS iterative --- ")
while queue:
    node = queue.popleft()
    print(node)
    for nei in D[node]:
        if nei not in seen:
            queue.append(nei)
    seen.add(node)
```

```python
from collections import defaultdict, deque
from typing import List


def can_reach_destination(start:int,end:int,bridges:List[tuple[int]]):
    if start == end:
        return (True,0)
    graph = defaultdict(list)
    for u,v in bridges:
        graph[u].append(v)
        graph[v].append(u)

    seen = {}
    q = deque()
    q.append((start,0))
    seen[start] = True
    while q:
        node,dist = q.popleft()
        for nei in graph[node]:
            if nei == end:
                return (True,dist + 1)
            if nei not in seen:
                q.append((nei,dist + 1))
                seen[node] = True

    return (False,-1)

bridges = [[1, 2], [2, 3], [3, 4]]
testcases = [
    (1,4),
    (1,5)
]

for testcase in testcases:
    print(testcase,can_reach_destination(testcase[0],testcase[1],bridges))

    
```
